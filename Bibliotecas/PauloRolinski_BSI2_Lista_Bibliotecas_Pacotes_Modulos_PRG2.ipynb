7. Importe a biblioteca matplotlib e crie um gráfico de barras simples com alguns
dados fictícios.

- Tabela de exemplo(imagem): Gráfico_de_barra_2+pt-BR.png

````python
import matplotlib.pyplot as plt
import numpy as np

species = ("Doce", "Lasanha", "Pizza", "Recheado", "Outros")
penguin_means = {
    'Lucro': (421995.04, 143760.97, 2800900.02, 443398.32, 384993.71),
    'Quantidade': (950184.03, 3000143, 3497265, 208304, 600689),
    'Custo': (602357.69, 120831.54, 2153942.33, 639807.59, 950540.50),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Dinheiro (em milhões)')
ax.set_title('Produtos(Top 4 do ano anterior por Lucro)')
ax.set_xticks(x + width, species)
ax.legend(loc='upper center', ncols=3)
ax.set_ylim(0, 4500000.00)

plt.show()
````

9. Importe o pacote requests e faça uma requisição HTTP para um URL e exiba o
código de status. HTTP_Status_Codes.webp

````python
import requests

urls = ['https://httpbin.org/basic-auth/user/pass', 'https://httpbin.org', 'https://canaltech.com.br/internet/sla/']

for i in urls:
  requisicao = requests.get(i)
  status = requisicao.status_code
  if status == 401:
    print(status, '-', 'Não autorizado')
  elif status == 200:
    print(status, '-', 'Ok')
  elif status == 404:
    print(status, '-', 'Não encontrado')
````
